```python
import seaborn as sns
sns.set_context("paper")
sns.set(font_scale = 1.1)
sns.set_style("white", {
    "font.family": "sans-serif",
    "font.serif": ['Helvetica'],
    "font.scale": 2
})
sns.set_style("ticks", {"xtick.major.size": 4,
                        "ytick.major.size": 4})

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

had_link_info_noAA = pd.read_pickle("data/had_link_info_noAA.pkl")
had_link_info = pd.read_pickle("data/had_link_info.pkl")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(3.5, 3.5))

had_link_info1h=had_link_info[(had_link_info.prev_load<3600*12)&(had_link_info.prev_load>0)]
had_link_info_noAA=had_link_info_noAA[(had_link_info_noAA.prev_load<3600*12)&(had_link_info_noAA.prev_load>0)]


plt.plot(had_link_info1h.prev_load, had_link_info1h.perc, alpha=0.4, linestyle=":", color="steelblue")
plt.plot(had_link_info_noAA.prev_load, had_link_info_noAA.perc, alpha=0.6, color="steelblue")

plt.xlabel("Interevent time AB (Log)")
plt.ylabel("Pages A with link to B (%)")
# plt.title("Referrer = Search engine")

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().set_ylim((0,0.7))
ax.set_xscale('log')
plt.xticks([1, 10, 60,  60*10, 3600, 12*3600], ["1s", "10s", '1m', "20m", "1h", "12h"]);

plt.savefig('out_in_from_search.pdf', bbox_inches='tight')
```
